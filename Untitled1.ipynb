# 4.2. Trực quan hóa dữ liệu

## Nhập thư viện, thiết lập cấu hình biểu đồ và đọc dữ liệu

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

#Thiết lập cấu hình biểu đồ
plt.rcParams['figure.figsize'] = (16, 10)
plt.rcParams['font.size'] = 13
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 12

# Đọc dữ liệu
df = pd.read_csv('financials_features_scaled.csv')
```

## 4.2.1. Biểu đồ Ma trận rủi ro - lợi nhuận
Dùng để so sánh mức sinh lời và mức độ rủi ro giữa các công ty. Giúp nhận biết doanh nghiệp nào có lợi nhuận cao nhưng rủi ro thấp.

Cách làm:
- Dùng plt.scatter() với các biến:
  + Trục X: price_to_52w_high__scaled
  + Trục Y: profitability_ratio__scaled
  + Kích thước: market_cap
  + Màu sắc: dividend_yield
- Dữ liệu được chuẩn hóa bằng StandardScaler() trước khi vẽ.
- Cấu hình thẩm mỹ: alpha=0.6, figsize=(16,12).

```python
fig, ax = plt.subplots(figsize=(16, 12))

# Chuẩn bị dữ liệu
rr = df[['price_to_52w_high__scaled', 'profitability_ratio__scaled', 
         'market_cap', 'dividend_yield']]

# Vẽ scatter plot
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(50, 1000))  # Bubble size min=50, max=1000
sizes = scaler.fit_transform(rr[['market_cap']])

scatter = ax.scatter(
    rr['price_to_52w_high__scaled'],
    rr['profitability_ratio__scaled'],
    c=rr['dividend_yield'],
    s=sizes,
    alpha=0.6,
    cmap='RdYlGn',
    edgecolors='black',
    linewidths=0.5
)

# Vẽ đường chia quadrant
ax.axhline(0, color='black', linestyle='-', linewidth=2, alpha=0.3)
ax.axvline(0, color='black', linestyle='-', linewidth=2, alpha=0.3)

# Thêm vùng màu nền
ax.axhspan(0, ax.get_ylim()[1], alpha=0.1, color='green', label='High Return')
ax.axhspan(ax.get_ylim()[0], 0, alpha=0.1, color='red')
ax.axvspan(ax.get_xlim()[0], 0, alpha=0.08, color='green')
ax.axvspan(0, ax.get_xlim()[1], alpha=0.08, color='red', label='High Risk')

# Labels và title
ax.set_xlabel('Rủi ro (Risk Score - Volatility)', fontsize=16, fontweight='bold')
ax.set_ylabel('Lợi nhuận (Return Score - Profitability)', fontsize=16, fontweight='bold')
ax.set_title('BIỂU ĐỒ PHÂN TÁN RỦI RO - LỢI NHUẬN\n(Kích thước bubble = Vốn hóa | Màu = Cổ tức)', 
             fontsize=18, fontweight='bold', pad=20)

# Thêm annotation cho các vùng
ax.text(0.02, 0.98, 'VÙNG LÝ TƯỞNG\nReturn cao, Risk thấp\n(Đầu tư tốt)', 
        transform=ax.transAxes, fontsize=14, verticalalignment='top',
        bbox=dict(boxstyle='round,pad=1', facecolor='lightgreen', alpha=0.8, edgecolor='darkgreen', linewidth=2),
        fontweight='bold')

ax.text(0.98, 0.02, 'VÙNG RỦI RO\nReturn thấp, Risk cao\n(Tránh đầu tư)', 
        transform=ax.transAxes, fontsize=14, verticalalignment='bottom', horizontalalignment='right',
        bbox=dict(boxstyle='round,pad=1', facecolor='lightcoral', alpha=0.8, edgecolor='darkred', linewidth=2),
        fontweight='bold')

# Colorbar
cbar = plt.colorbar(scatter, ax=ax, pad=0.02)
cbar.set_label('Dividend Yield (%)', fontsize=14, fontweight='bold')

ax.grid(True, alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()
```

## 4.2.2. Biểu đồ Phân phối EBITDA
Cho thấy phân bố giá trị EBITDA (lợi nhuận trước thuế, lãi vay và khấu hao) của các công ty. Giúp xem EBITDA có bị lệch, phân tán hay tập trung quanh một mức nhất định.
Dùng biến ebitda_log__scaled, phiên bản ổn định hơn của EBITDA, hỗ trợ so sánh công bằng.

Cách làm:
- Dùng histogram thể hiện phân phối của biến ebitda_log__scaled (log chuẩn hóa của EBITDA).
- Chỉ lấy các giá trị hợp lệ (loại bỏ ngoại lệ quá thấp).
- Thêm đường trung bình (màu đỏ) và trung vị (màu xanh)** để so sánh xu hướng phân phối.
- Màu chủ đạo xanh lá, viền đen, độ trong suốt trung bình giúp nổi bật dạng phân bố.

```python
plt.figure(figsize=(14, 8))
ebitda_positive = df[df['ebitda_log__scaled'] > -5]['ebitda_log__scaled']

plt.hist(ebitda_positive, bins=60, color='#2ECC71', edgecolor='black', 
         alpha=0.7, linewidth=1.2)
plt.axvline(ebitda_positive.mean(), color='red', linestyle='--', 
            linewidth=2, label=f'Trung bình: {ebitda_positive.mean():.2f}')
plt.axvline(ebitda_positive.median(), color='blue', linestyle='--', 
            linewidth=2, label=f'Trung vị: {ebitda_positive.median():.2f}')

plt.xlabel('EBITDA (log scaled)', fontsize=14, fontweight='bold')
plt.ylabel('Số lượng công ty', fontsize=14, fontweight='bold')
plt.title('PHÂN PHỐI EBITDA - HIỆU QUẢ HOẠT ĐỘNG', fontsize=18, fontweight='bold', pad=20)
plt.legend(fontsize=12, loc='best')
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
```

## 4.2.3: Ma trận tương quan (Heatmap)
Biểu diễn mối tương quan giữa các biến tài chính: P/E, P/B, P/S, tỉ lệ cổ tức, lợi nhuận, vốn hóa và EBITDA. Giúp phát hiện biến nào liên quan chặt chẽ, hỗ trợ cho mô hình phân tích.

Cách làm:
- Tính ma trận tương quan (correlation matrix) giữa các biến tài chính quan trọng: price_earnings, price_book, price_sales, dividend_yield, profitability_ratio__scaled, market_cap_log__scaled, và ebitda_log__scaled.
- Dùng heatmap (bản đồ nhiệt) để trực quan hóa mức độ tương quan, hiển thị giá trị hệ số tương quan (từ -1 đến +1) ngay trên từng ô.
- Thang màu “coolwarm” được sử dụng:bMàu đỏ đậm biểu thị tương quan dương mạnh (các biến cùng tăng hoặc cùng giảm), màu xanh đậm biểu thị tương quan âm mạnh (một biến tăng, biến kia giảm).
- Bố cục được căn chỉnh vuông, có khung trắng giữa các ô và tiêu đề giải thích rõ ý nghĩa.

```python
fig, ax = plt.subplots(figsize=(14, 11))

# Chọn các chỉ số quan trọng
corr_cols = ['price_earnings', 'price_book', 'price_sales', 'dividend_yield', 
             'profitability_ratio__scaled', 'market_cap_log__scaled', 'ebitda_log__scaled']

corr_labels = ['P/E Ratio', 'P/B Ratio', 'P/S Ratio', 'Dividend Yield',
               'Profitability', 'Market Cap (log)', 'EBITDA (log)']

corr_matrix = df[corr_cols].corr()

# Vẽ heatmap
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0, 
            ax=ax, cbar_kws={'shrink': 0.8}, linewidths=2, linecolor='white',
            square=True, annot_kws={'size': 13, 'weight': 'bold'})

# Đặt labels
ax.set_xticklabels(corr_labels, rotation=45, ha='right', fontsize=12)
ax.set_yticklabels(corr_labels, rotation=0, fontsize=12)
ax.set_title('MA TRẬN TƯƠNG QUAN CÁC CHỈ SỐ TÀI CHÍNH\n(Giá trị càng gần ±1 = tương quan càng mạnh)', 
             fontsize=18, fontweight='bold', pad=20)

plt.tight_layout()
plt.show()
```

## 4.2.4. Biểu đồ Ảnh hưởng của vốn hóa lên hiệu qủa hoạt động
Thể hiện mối quan hệ giữa quy mô doanh nghiệp (vốn hóa thị trường) và hiệu quả hoạt động (EBITDA). Mục đích là xem doanh nghiệp lớn có hoạt động hiệu quả hơn hay không.

Cách làm: Vẽ 2 biểu đồ đặt cạnh nhau (subplot) để thể hiện tác động của quy mô vốn hóa thị trường đến hiệu quả hoạt động tài chính.
- Bubble chart: Thể hiện mối quan hệ giữa EBITDA (log scaled) và Tỷ lệ sinh lời (profitability_ratio__scaled).
  + Chia doanh nghiệp thành 3 nhóm theo quy mô vốn hóa (Nhỏ, Trung bình, Lớn) bằng phương pháp phân loại theo khoảng giá trị (pd.cut).
  + Kích thước bong bóng thể hiện tỷ suất cổ tức (dividend yield); màu sắc khác nhau cho từng nhóm quy mô.
  + Thêm đường trục ngang và dọc (0,0) để phân tách các vùng hoạt động.
- Bar chart: so sánh giá trị trung bình của các chỉ tiêu chính (Sinh lời, Cổ tức, Định giá) giữa ba nhóm quy mô.
  + Mỗi màu đại diện cho một nhóm (Nhỏ – đỏ, Trung bình – cam, Lớn – xanh).
  + Các cột được sắp xếp cạnh nhau để dễ nhận thấy sự khác biệt.

```python
fig = plt.figure(figsize=(18, 8))
gs = fig.add_gridspec(1, 2, wspace=0.3)

# Biểu đồ 1: Bubble chart vốn hóa vs các chỉ số
ax1 = fig.add_subplot(gs[0])

df['size_group'] = pd.cut(df['market_cap_log__scaled'], 
                          bins=[-np.inf, -0.5, 0.5, np.inf],
                          labels=['Nhỏ', 'Trung bình', 'Lớn'])

size_colors = {'Nhỏ': '#E74C3C', 'Trung bình': '#F39C12', 'Lớn': '#27AE60'}

for group in ['Nhỏ', 'Trung bình', 'Lớn']:
    group_data = df[df['size_group'] == group][
        ['ebitda_log__scaled', 'profitability_ratio__scaled', 'dividend_yield']
    ].dropna()
    
    group_data = group_data[(group_data['ebitda_log__scaled'] > -5)]
    
    ax1.scatter(group_data['ebitda_log__scaled'],
               group_data['profitability_ratio__scaled'],
               s=np.abs(group_data['dividend_yield']) * 100 + 20,
               alpha=0.6, label=f'Vốn hóa {group}',
               color=size_colors[group], edgecolors='black', linewidth=0.5)

ax1.axhline(y=0, color='red', linestyle='--', linewidth=2, alpha=0.5)
ax1.axvline(x=0, color='blue', linestyle='--', linewidth=2, alpha=0.5)

ax1.set_xlabel('EBITDA (log scaled)', fontsize=13, fontweight='bold')
ax1.set_ylabel('Tỷ lệ sinh lời', fontsize=13, fontweight='bold')
ax1.set_title('EBITDA × Sinh lời theo quy mô\n(Kích thước = Tỷ suất cổ tức)', 
              fontsize=14, fontweight='bold', pad=15)
ax1.legend(loc='upper left', fontsize=11, framealpha=0.9)
ax1.grid(True, alpha=0.3)

# Biểu đồ 2: so sánh theo vốn hóa
ax2 = fig.add_subplot(gs[1])

# Chuẩn bị dữ liệu trung bình theo nhóm và chỉ tiêu
metrics = [
    ('profitability_ratio__scaled', 'Sinh lời'),
    ('dividend_yield', 'Cổ tức %'),
    ('valuation_z__scaled', 'Định giá')
]

bar_data = []

for metric, label in metrics:
    mean_values = (
        df.groupby('size_group')[metric]
        .mean()
        .reset_index()
        .rename(columns={metric: 'Giá trị trung bình'})
    )
    mean_values['Chỉ tiêu'] = label
    bar_data.append(mean_values)

bar_df = pd.concat(bar_data, ignore_index=True)

# Vẽ bar chart đúng
sns.barplot(
    data=bar_df,
    x='Chỉ tiêu',
    y='Giá trị trung bình',
    hue='size_group',
    palette=size_colors,
    errorbar='sd',
    ax=ax2
)

# Cấu hình biểu đồ
ax2.set_ylabel('Giá trị trung bình (scaled)', fontsize=13, fontweight='bold')
ax2.set_xlabel('')
ax2.set_title('So sánh trung bình các chỉ số theo quy mô', fontsize=14, fontweight='bold', pad=15)
ax2.grid(True, alpha=0.3, axis='y')
ax2.legend(title='Quy mô', fontsize=10, title_fontsize=11, framealpha=0.9)
plt.setp(ax2.xaxis.get_majorticklabels(), rotation=0, fontsize=11)

plt.suptitle('ẢNH HƯỞNG CỦA VỐN HÓA LÊN HIỆU QUẢ HOẠT ĐỘNG', 
             fontsize=16, fontweight='bold', y=0.98)

plt.tight_layout()
plt.show()
```

## 4.2.5. Biểu đồ So sánh các công ty có/không có cổ tức

Dùng để so sánh đặc điểm tài chính và tỷ trọng doanh nghiệp giữa hai nhóm trong bộ dữ liệu: Nhóm có chi trả cổ tức và nhóm không chi trả cổ tức. Mục tiêu là xác định sự khác biệt về hiệu quả hoạt động, quy mô, định giá và tần suất xuất hiện giữa hai loại hình doanh nghiệp này.

Cách làm: Dùng 2 biểu đồ con (subplot) hiển thị song song trong cùng khung hình, bố trí bằng plt.subplots(1, 2, figsize=(16, 8)).
- Biểu đồ cột (bar chart)
    + So sánh giá trị trung bình của ba chỉ số tài chính: Tỷ lệ sinh lời (profitability_ratio__scaled), Vốn hóa thị trường (market_cap_log__scaled) và Định giá tổng hợp (valuation_z__scaled).
    + Chia thành hai nhóm doanh nghiệp dựa trên biến dividend_flag__scaled: Có cổ tức (0) – màu xanh và Không cổ tức (-1) – màu đỏ.
    + Các cột đặt cạnh nhau giúp dễ so sánh giữa hai nhóm.
- Biểu đồ 2 (bên phải): Biểu đồ tròn (pie chart)
    + Thể hiện tỷ lệ phần trăm và số lượng công ty thuộc hai nhóm “có cổ tức” và “không cổ tức”, màu xanh biểu thị nhóm có cổ tức, màu đỏ cho nhóm không cổ tức.
    + Thêm nhãn hiển thị phần trăm và số lượng công ty ngay trên biểu đồ.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

# Chart 1: So sánh các chỉ số trung bình
dividend_comparison = df.groupby('dividend_flag__scaled')[
    ['profitability_ratio__scaled', 'market_cap_log__scaled', 'valuation_z__scaled']
].mean()

x = np.arange(len(dividend_comparison.columns))
width = 0.35
labels = ['Không cổ tức (-1)', 'Có cổ tức (0)']
colors_bars = ['#E74C3C', '#3498DB']

for idx, (label, color) in enumerate(zip(labels, colors_bars)):
    values = dividend_comparison.iloc[idx].values
    ax1.bar(x + idx*width, values, width, label=label, color=color, alpha=0.8)

ax1.set_xlabel('Chỉ số tài chính', fontsize=12, fontweight='bold')
ax1.set_ylabel('Giá trị trung bình', fontsize=12, fontweight='bold')
ax1.set_title('So sánh chỉ số trung bình', fontsize=14, fontweight='bold')
ax1.set_xticks(x + width/2)
ax1.set_xticklabels(['Sinh lời', 'Vốn hóa', 'Định giá'])
ax1.legend()
ax1.grid(True, alpha=0.3, axis='y')

# Chart 2: Số lượng công ty
dividend_counts = df['dividend_flag__scaled'].value_counts()
colors_pie = ['#3498DB', '#E74C3C']
labels_pie = [f'Có cổ tức\n({dividend_counts[0]} công ty)', 
              f'Không cổ tức\n({dividend_counts[-1]} công ty)']

ax2.pie(dividend_counts.values, labels=labels_pie, colors=colors_pie,
        autopct='%1.1f%%', startangle=90, textprops={'fontsize': 12, 'fontweight': 'bold'})
ax2.set_title('Tỷ lệ công ty có/không cổ tức', fontsize=14, fontweight='bold')

plt.suptitle('PHÂN TÍCH CÔNG TY CÓ/KHÔNG CỔ TỨC', 
             fontsize=18, fontweight='bold', y=0.98)
plt.tight_layout()
plt.show()
```
