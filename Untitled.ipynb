```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Cấu hình
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (20, 12)
plt.rcParams['font.size'] = 10

# Đọc dữ liệu
df = pd.read_csv('financials_features_scaled.csv')
print(f"\n✓ Đã tải: {df.shape[0]} công ty, {df.shape[1]} đặc trưng\n")

# 1. PHÂN TÍCH TỔNG QUAN

fig = plt.figure(figsize=(20, 12))

# 1.1 P/E Ratio
ax1 = plt.subplot(3, 3, 1)
pe = df[(df['price_earnings'] > 0) & (df['price_earnings'] < 100)]['price_earnings']
ax1.hist(pe, bins=30, color='steelblue', edgecolor='black', alpha=0.7)
ax1.axvline(pe.median(), color='red', linestyle='--', label=f'Median: {pe.median():.1f}')
ax1.set_xlabel('P/E Ratio')
ax1.set_ylabel('Số công ty')
ax1.set_title('Phân phối P/E Ratio', fontweight='bold', fontsize=12)
ax1.legend()
ax1.grid(alpha=0.3)

# 1.2 Market Cap
ax2 = plt.subplot(3, 3, 2)
cap_b = df['market_cap'] / 1e9
ax2.hist(np.log10(cap_b[cap_b > 0]), bins=30, color='coral', edgecolor='black', alpha=0.7)
ax2.set_xlabel('log10(Market Cap - Tỷ USD)')
ax2.set_ylabel('Số công ty')
ax2.set_title('Phân phối Vốn hóa thị trường', fontweight='bold', fontsize=12)
ax2.grid(alpha=0.3)

# 1.3 Dividend Yield
ax3 = plt.subplot(3, 3, 3)
div = df[df['dividend_yield'] > 0]['dividend_yield']
ax3.hist(div, bins=30, color='green', edgecolor='black', alpha=0.7)
ax3.axvline(div.median(), color='red', linestyle='--', label=f'Median: {div.median():.2f}%')
ax3.set_xlabel('Dividend Yield (%)')
ax3.set_ylabel('Số công ty')
ax3.set_title(f'Cổ tức ({len(div)}/{len(df)} công ty)', fontweight='bold', fontsize=12)
ax3.legend()
ax3.grid(alpha=0.3)

# 1.4 P/E: Có/không cổ tức
ax4 = plt.subplot(3, 3, 4)
df['has_div'] = df['dividend_yield'] > 0
pe_comp = df[(df['price_earnings'] > 0) & (df['price_earnings'] < 100)]
sns.boxplot(data=pe_comp, x='has_div', y='price_earnings', ax=ax4)
ax4.set_xticklabels(['Không trả cổ tức', 'Có trả cổ tức'])
ax4.set_ylabel('P/E Ratio')
ax4.set_title('So sánh P/E: Growth vs Value', fontweight='bold', fontsize=12)
ax4.grid(alpha=0.3, axis='y')

# 1.5 Phân loại theo Market Cap
ax5 = plt.subplot(3, 3, 5)
def cat_cap(x):
    b = x / 1e9
    if b > 200: return 'Mega (>200B)'
    elif b >= 10: return 'Large (10-200B)'
    elif b >= 2: return 'Mid (2-10B)'
    else: return 'Small (<2B)'

df['cap_cat'] = df['market_cap'].apply(cat_cap)
cap_c = df['cap_cat'].value_counts()
ax5.pie(cap_c, labels=cap_c.index, autopct='%1.1f%%', startangle=90,
        colors=['#8b5cf6', '#3b82f6', '#10b981', '#f59e0b'])
ax5.set_title('Phân loại theo quy mô', fontweight='bold', fontsize=12)

# 1.6 Top 15 công ty
ax6 = plt.subplot(3, 3, 6)
top15 = df.nlargest(15, 'market_cap').copy()
top15['cap_b'] = top15['market_cap'] / 1e9
top15['name'] = [f'C{i+1}' for i in range(len(top15))]
ax6.barh(range(len(top15)), top15['cap_b'], color='purple', alpha=0.7)
ax6.set_yticks(range(len(top15)))
ax6.set_yticklabels(top15['name'], fontsize=9)
ax6.set_xlabel('Market Cap (Tỷ USD)')
ax6.set_title('Top 15 công ty lớn nhất', fontweight='bold', fontsize=12)
ax6.grid(alpha=0.3, axis='x')

# 1.7 P/B Ratio
ax7 = plt.subplot(3, 3, 7)
pb = df[(df['price_book'] > 0) & (df['price_book'] < 20)]['price_book']
ax7.hist(pb, bins=30, color='indianred', edgecolor='black', alpha=0.7)
ax7.axvline(pb.median(), color='blue', linestyle='--', label=f'Median: {pb.median():.2f}')
ax7.set_xlabel('P/B Ratio')
ax7.set_ylabel('Số công ty')
ax7.set_title('Price-to-Book Ratio', fontweight='bold', fontsize=12)
ax7.legend()
ax7.grid(alpha=0.3)

# 1.8 Ma trận tương quan
ax8 = plt.subplot(3, 3, 8)
corr_cols = ['price_earnings', 'price_book', 'dividend_yield', 
             'profitability_ratio__scaled', 'market_cap_log__scaled']
corr = df[corr_cols].corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0, 
            ax=ax8, cbar_kws={'shrink': 0.8})
ax8.set_title('Ma trận tương quan', fontweight='bold', fontsize=12)

# 1.9 EBITDA
ax9 = plt.subplot(3, 3, 9)
ebitda = df[df['ebitda'] > 0]['ebitda'] / 1e9
ebitda = ebitda[ebitda < ebitda.quantile(0.95)]
ax9.hist(ebitda, bins=30, color='teal', edgecolor='black', alpha=0.7)
ax9.set_xlabel('EBITDA (Tỷ USD)')
ax9.set_ylabel('Số công ty')
ax9.set_title('Phân phối EBITDA', fontweight='bold', fontsize=12)
ax9.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# 2. PHÂN TÍCH ĐỊNH GIÁ

fig = plt.figure(figsize=(20, 10))

# 2.1 P/E vs Dividend
ax1 = plt.subplot(2, 3, 1)
valid = df[(df['price_earnings'] > 0) & (df['price_earnings'] < 80)]
scatter = ax1.scatter(valid['price_earnings'], valid['dividend_yield'],
                     c=valid['market_cap_log__scaled'], cmap='viridis',
                     s=50, alpha=0.6, edgecolors='black')
ax1.set_xlabel('P/E Ratio')
ax1.set_ylabel('Dividend Yield (%)')
ax1.set_title('P/E vs Dividend Yield', fontweight='bold', fontsize=12)
plt.colorbar(scatter, ax=ax1, label='Market Cap')
ax1.grid(alpha=0.3)

# 2.2 P/E vs P/B
ax2 = plt.subplot(2, 3, 2)
valid2 = df[(df['price_earnings'] > 0) & (df['price_earnings'] < 80) &
            (df['price_book'] > 0) & (df['price_book'] < 20)]
ax2.scatter(valid2['price_earnings'], valid2['price_book'], 
           alpha=0.5, s=50, edgecolors='black')
z = np.polyfit(valid2['price_earnings'], valid2['price_book'], 1)
p = np.poly1d(z)
ax2.plot(valid2['price_earnings'], p(valid2['price_earnings']),
         "r--", alpha=0.8, label=f'y={z[0]:.2f}x+{z[1]:.2f}')
ax2.set_xlabel('P/E Ratio')
ax2.set_ylabel('P/B Ratio')
ax2.set_title('P/E vs P/B', fontweight='bold', fontsize=12)
ax2.legend()
ax2.grid(alpha=0.3)

# 2.3 Phân loại P/E
ax3 = plt.subplot(2, 3, 3)
def cat_pe(x):
    if x <= 0: return 'Âm'
    elif x < 15: return 'Thấp (<15)'
    elif x < 25: return 'TB (15-25)'
    elif x < 40: return 'Cao (25-40)'
    else: return 'Rất cao (>40)'

df['pe_cat'] = df['price_earnings'].apply(cat_pe)
pe_c = df['pe_cat'].value_counts()
bars = ax3.bar(range(len(pe_c)), pe_c.values,
               color=['#ef4444', '#10b981', '#3b82f6', '#f59e0b', '#8b5cf6'])
ax3.set_xticks(range(len(pe_c)))
ax3.set_xticklabels(pe_c.index, rotation=45, ha='right')
ax3.set_ylabel('Số công ty')
ax3.set_title('Phân loại theo P/E', fontweight='bold', fontsize=12)
ax3.grid(alpha=0.3, axis='y')

# 2.4 Price/Sales
ax4 = plt.subplot(2, 3, 4)
ps = df[(df['price_sales'] > 0) & (df['price_sales'] < 20)]['price_sales']
ax4.hist(ps, bins=30, color='gold', edgecolor='black', alpha=0.7)
ax4.axvline(ps.median(), color='red', linestyle='--', label=f'Median: {ps.median():.2f}')
ax4.set_xlabel('P/S Ratio')
ax4.set_ylabel('Số công ty')
ax4.set_title('Price-to-Sales', fontweight='bold', fontsize=12)
ax4.legend()
ax4.grid(alpha=0.3)

# 2.5 Valuation Score
ax5 = plt.subplot(2, 3, 5)
df['val_score'] = 0
df.loc[df['price_earnings'].between(0, 15), 'val_score'] += 3
df.loc[df['price_earnings'].between(15, 25), 'val_score'] += 2
df.loc[df['price_book'] < 3, 'val_score'] += 2
df.loc[df['dividend_yield'] > 2, 'val_score'] += 2

val_dist = df['val_score'].value_counts().sort_index()
ax5.bar(val_dist.index, val_dist.values,
        color=plt.cm.RdYlGn(val_dist.index / 7), edgecolor='black')
ax5.set_xlabel('Valuation Score (0-7)')
ax5.set_ylabel('Số công ty')
ax5.set_title('Điểm định giá tổng hợp', fontweight='bold', fontsize=12)
ax5.grid(alpha=0.3, axis='y')

# 2.6 Top undervalued
ax6 = plt.subplot(2, 3, 6)
top_val = df[df['val_score'] >= 5].nlargest(10, 'market_cap')
if len(top_val) > 0:
    ax6.barh(range(len(top_val)), top_val['val_score'],
             color=plt.cm.Greens(np.linspace(0.4, 0.9, len(top_val))))
    ax6.set_yticks(range(len(top_val)))
    ax6.set_yticklabels([f'C{i+1}' for i in range(len(top_val))], fontsize=9)
    ax6.set_xlabel('Valuation Score')
    ax6.set_title('Top 10 định giá hấp dẫn', fontweight='bold', fontsize=12)
    ax6.grid(alpha=0.3, axis='x')

plt.tight_layout()
plt.show()

# 3. RỦI RO - LỢI NHUẬN

fig = plt.figure(figsize=(20, 10))

# 3.1 Ma trận Rủi ro-Lợi nhuận
ax1 = plt.subplot(2, 3, 1)
rr = df[['price_to_52w_high__scaled', 'profitability_ratio__scaled', 
         'market_cap', 'dividend_yield']].dropna()

scatter = ax1.scatter(rr['price_to_52w_high__scaled'],
                     rr['profitability_ratio__scaled'],
                     c=rr['dividend_yield'], s=np.sqrt(rr['market_cap'])/1e5,
                     alpha=0.6, cmap='RdYlGn', edgecolors='black')
ax1.axhline(0, color='gray', linestyle='--', alpha=0.5)
ax1.axvline(0, color='gray', linestyle='--', alpha=0.5)
ax1.set_xlabel('Risk (volatility)')
ax1.set_ylabel('Return (profitability)')
ax1.set_title('MA TRẬN RỦI RO - LỢI NHUẬN', fontweight='bold', fontsize=13)
plt.colorbar(scatter, ax=ax1, label='Dividend %')
ax1.grid(alpha=0.3)

# 3.2 Risk-Return Quadrants
ax2 = plt.subplot(2, 3, 2)
def classify(row):
    r = row['price_to_52w_high__scaled']
    ret = row['profitability_ratio__scaled']
    if ret > 0 and r < 0: return 'Tốt (High R, Low Risk)'
    elif ret > 0: return 'TB+ (High R, High Risk)'
    elif r < 0: return 'TB- (Low R, Low Risk)'
    else: return 'Kém (Low R, High Risk)'

rr['quad'] = rr.apply(classify, axis=1)
quad_c = rr['quad'].value_counts()
ax2.pie(quad_c, labels=quad_c.index, autopct='%1.1f%%', startangle=90,
        colors=['#10b981', '#3b82f6', '#f59e0b', '#ef4444'])
ax2.set_title('Phân loại Risk-Return', fontweight='bold', fontsize=12)

# 3.3 Volatility
ax3 = plt.subplot(2, 3, 3)
vol = df['52w_range__scaled'].dropna()
ax3.hist(vol, bins=30, color='orange', edgecolor='black', alpha=0.7)
ax3.axvline(vol.median(), color='red', linestyle='--', label=f'Median: {vol.median():.2f}')
ax3.set_xlabel('52-week Range (scaled)')
ax3.set_ylabel('Số công ty')
ax3.set_title('Phân phối Volatility', fontweight='bold', fontsize=12)
ax3.legend()
ax3.grid(alpha=0.3)

# 3.4 Profitability
ax4 = plt.subplot(2, 3, 4)
prof = df['profitability_ratio__scaled'].dropna()
ax4.hist(prof, bins=40, color='green', edgecolor='black', alpha=0.7)
ax4.axvline(0, color='red', linestyle='--', linewidth=2, label='Breakeven')
ax4.axvline(prof.median(), color='blue', linestyle='--', label=f'Median: {prof.median():.2f}')
ax4.set_xlabel('Profitability (scaled)')
ax4.set_ylabel('Số công ty')
ax4.set_title('Khả năng sinh lời', fontweight='bold', fontsize=12)
ax4.legend()
ax4.grid(alpha=0.3)

# 3.5 EPS
ax5 = plt.subplot(2, 3, 5)
eps = df[df['earnings_share'] > 0]['earnings_share']
eps = eps[eps < eps.quantile(0.95)]
ax5.hist(eps, bins=30, color='purple', edgecolor='black', alpha=0.7)
ax5.axvline(eps.median(), color='red', linestyle='--', label=f'Median: ${eps.median():.2f}')
ax5.set_xlabel('EPS ($)')
ax5.set_ylabel('Số công ty')
ax5.set_title('Earnings Per Share', fontweight='bold', fontsize=12)
ax5.legend()
ax5.grid(alpha=0.3)

# 3.6 Risk-Adjusted Return
ax6 = plt.subplot(2, 3, 6)
df['radj'] = df['profitability_ratio__scaled'] / (abs(df['52w_range__scaled']) + 0.1)
radj = df['radj'].dropna()
radj = radj[(radj > radj.quantile(0.05)) & (radj < radj.quantile(0.95))]
ax6.hist(radj, bins=30, color='teal', edgecolor='black', alpha=0.7)
ax6.axvline(radj.median(), color='red', linestyle='--', label=f'Median: {radj.median():.2f}')
ax6.set_xlabel('Risk-Adjusted Return')
ax6.set_ylabel('Số công ty')
ax6.set_title('Lợi nhuận điều chỉnh rủi ro', fontweight='bold', fontsize=12)
ax6.legend()
ax6.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# 4. SỨC KHỎE TÀI CHÍNH

fig = plt.figure(figsize=(20, 10))

# Tính điểm
df['health'] = 0
df.loc[df['profitability_ratio__scaled'] > 1, 'health'] += 40
df.loc[df['profitability_ratio__scaled'].between(0, 1), 'health'] += 30
df.loc[df['dividend_yield'] > 2, 'health'] += 10
df.loc[df['price_earnings'].between(10, 25), 'health'] += 15
df.loc[df['price_book'] < 3, 'health'] += 15
df.loc[df['52w_range__scaled'] < 0, 'health'] += 10
df.loc[df['ebitda'] > 0, 'health'] += 10

# 4.1 Top 30
ax1 = plt.subplot(2, 2, 1)
top30 = df.nlargest(30, 'health')
ax1.barh(range(len(top30)), top30['health'],
         color=plt.cm.RdYlGn(top30['health'] / 100))
ax1.set_yticks(range(len(top30)))
ax1.set_yticklabels([f'C{i+1}' for i in range(len(top30))], fontsize=8)
ax1.set_xlabel('Health Score (0-100)')
ax1.set_title('TOP 30 SỨC KHỎE TỐT NHẤT', fontweight='bold', fontsize=13)
ax1.axvline(70, color='green', linestyle='--', alpha=0.5, label='Tốt (70)')
ax1.axvline(50, color='orange', linestyle='--', alpha=0.5, label='TB (50)')
ax1.legend()
ax1.grid(alpha=0.3, axis='x')

# 4.2 Phân phối
ax2 = plt.subplot(2, 2, 2)
ax2.hist(df['health'], bins=20, color='royalblue', edgecolor='black', alpha=0.7)
ax2.axvline(df['health'].median(), color='red', linestyle='--',
            label=f'Median: {df["health"].median():.1f}')
ax2.axvline(70, color='green', linestyle='--', alpha=0.7)
ax2.axvline(50, color='orange', linestyle='--', alpha=0.7)
ax2.set_xlabel('Health Score')
ax2.set_ylabel('Số công ty')
ax2.set_title('Phân phối Health Score', fontweight='bold', fontsize=12)
ax2.legend()
ax2.grid(alpha=0.3)

# 4.3 Phân loại
ax3 = plt.subplot(2, 2, 3)
def cat_health(x):
    if x >= 70: return 'Xuất sắc (>=70)'
    elif x >= 50: return 'Tốt (50-69)'
    elif x >= 30: return 'Trung bình (30-49)'
    else: return 'Yếu (<30)'

df['health_cat'] = df['health'].apply(cat_health)
hc = df['health_cat'].value_counts()
ax3.pie(hc, labels=hc.index, autopct='%1.1f%%', startangle=90,
        colors=['#10b981', '#3b82f6', '#f59e0b', '#ef4444'])
ax3.set_title('Phân loại Health Score', fontweight='bold', fontsize=12)

# 4.4 Health vs Market Cap
ax4 = plt.subplot(2, 2, 4)
scatter = ax4.scatter(df['health'], np.log10(df['market_cap']/1e9),
                     c=df['dividend_yield'], cmap='plasma',
                     s=50, alpha=0.6, edgecolors='black')
ax4.set_xlabel('Health Score')
ax4.set_ylabel('log10(Market Cap - Tỷ USD)')
ax4.set_title('Health Score vs Quy mô', fontweight='bold', fontsize=12)
plt.colorbar(scatter, ax=ax4, label='Dividend %')
ax4.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# TÓM TẮT

print(f"\nTỔNG QUAN:")
print(f"  - Tổng số công ty: {len(df)}")
print(f"  - P/E trung bình: {df[df['price_earnings']>0]['price_earnings'].mean():.2f}")
print(f"  - Công ty trả cổ tức: {len(df[df['dividend_yield']>0])} ({len(df[df['dividend_yield']>0])/len(df)*100:.1f}%)")
print(f"  - Dividend TB: {df['dividend_yield'].mean():.2f}%")

print(f"\nĐỊNH GIÁ:")
print(f"  - P/E < 15 (value): {len(df[df['price_earnings'].between(0,15)])} công ty")
print(f"  - P/E 15-25 (fair): {len(df[df['price_earnings'].between(15,25)])} công ty")  
print(f"  - P/E > 25 (growth): {len(df[df['price_earnings']>25])} công ty")

print(f"\nSỨC KHỎE:")
print(f"  - Xuất sắc (>=70): {len(df[df['health']>=70])} công ty")
print(f"  - Tốt (50-69): {len(df[df['health'].between(50,69)])} công ty")
print(f"  - Trung bình (<50): {len(df[df['health']<50])} công ty")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Cấu hình FONT SIZE LỚN
plt.rcParams['figure.figsize'] = (16, 10)
plt.rcParams['font.size'] = 13
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 12

# Đọc dữ liệu
df = pd.read_csv('financials_features_scaled.csv')

# BIỂU ĐỒ 1: MA TRẬN RỦI RO - LỢI NHUẬN

fig, ax = plt.subplots(figsize=(16, 12))

# Chuẩn bị dữ liệu
rr = df[['price_to_52w_high__scaled', 'profitability_ratio__scaled', 
         'market_cap', 'dividend_yield']].dropna()

# Vẽ scatter plot
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(50, 1000))  # Bubble size min=50, max=1000
sizes = scaler.fit_transform(rr[['market_cap']])

scatter = ax.scatter(
    rr['price_to_52w_high__scaled'],
    rr['profitability_ratio__scaled'],
    c=rr['dividend_yield'],
    s=sizes,
    alpha=0.6,
    cmap='RdYlGn',
    edgecolors='black',
    linewidths=0.5
)

# Vẽ đường chia quadrant
ax.axhline(0, color='black', linestyle='-', linewidth=2, alpha=0.3)
ax.axvline(0, color='black', linestyle='-', linewidth=2, alpha=0.3)

# Thêm vùng màu nền
ax.axhspan(0, ax.get_ylim()[1], alpha=0.1, color='green', label='High Return')
ax.axhspan(ax.get_ylim()[0], 0, alpha=0.1, color='red')
ax.axvspan(ax.get_xlim()[0], 0, alpha=0.08, color='green')
ax.axvspan(0, ax.get_xlim()[1], alpha=0.08, color='red', label='High Risk')

# Labels và title
ax.set_xlabel('Rủi ro (Risk Score - Volatility)', fontsize=16, fontweight='bold')
ax.set_ylabel('Lợi nhuận (Return Score - Profitability)', fontsize=16, fontweight='bold')
ax.set_title('MA TRẬN RỦI RO - LỢI NHUẬN\n(Kích thước bubble = Vốn hóa | Màu = Cổ tức)', 
             fontsize=18, fontweight='bold', pad=20)

# Thêm annotation cho các vùng
ax.text(0.02, 0.98, 'VÙNG LÝ TƯỞNG\nReturn cao, Risk thấp\n(Đầu tư tốt)', 
        transform=ax.transAxes, fontsize=14, verticalalignment='top',
        bbox=dict(boxstyle='round,pad=1', facecolor='lightgreen', alpha=0.8, edgecolor='darkgreen', linewidth=2),
        fontweight='bold')

ax.text(0.98, 0.02, 'VÙNG RỦI RO\nReturn thấp, Risk cao\n(Tránh đầu tư)', 
        transform=ax.transAxes, fontsize=14, verticalalignment='bottom', horizontalalignment='right',
        bbox=dict(boxstyle='round,pad=1', facecolor='lightcoral', alpha=0.8, edgecolor='darkred', linewidth=2),
        fontweight='bold')

# Colorbar
cbar = plt.colorbar(scatter, ax=ax, pad=0.02)
cbar.set_label('Dividend Yield (%)', fontsize=14, fontweight='bold')

ax.grid(True, alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()

# BIỂU ĐỒ 2: ĐIỂM SỨC KHỎE TÀI CHÍNH - TOP 40

# Tính điểm sức khỏe
df['health'] = 0
df.loc[df['profitability_ratio__scaled'] > 1, 'health'] += 40
df.loc[df['profitability_ratio__scaled'].between(0, 1), 'health'] += 30
df.loc[df['dividend_yield'] > 2, 'health'] += 10
df.loc[df['dividend_yield'].between(0, 2), 'health'] += 5
df.loc[df['price_earnings'].between(10, 25), 'health'] += 15
df.loc[df['price_book'] < 3, 'health'] += 15
df.loc[df['52w_range__scaled'] < 0, 'health'] += 10
df.loc[df['ebitda'] > 0, 'health'] += 10

fig, ax = plt.subplots(figsize=(16, 14))

# Top 40 công ty
top40 = df.nlargest(40, 'health').reset_index(drop=True)
top40['name'] = [f'Công ty #{i+1}' for i in range(len(top40))]

# Màu gradient
colors = plt.cm.RdYlGn(top40['health'] / 100)

# Vẽ horizontal bar
bars = ax.barh(range(len(top40)), top40['health'], color=colors, 
               edgecolor='black', linewidth=1.2)

# Labels
ax.set_yticks(range(len(top40)))
ax.set_yticklabels(top40['name'], fontsize=11)
ax.set_xlabel('Điểm Sức khỏe Tài chính (0-100)', fontsize=16, fontweight='bold')
ax.set_title('TOP 40 CÔNG TY CÓ SỨC KHỎE TÀI CHÍNH TỐT NHẤT\n(Dựa trên Profitability, Valuation, Dividend, Stability)', 
             fontsize=18, fontweight='bold', pad=20)

# Thêm đường ngưỡng
ax.axvline(70, color='green', linestyle='--', linewidth=2.5, alpha=0.7, label='Ngưỡng Tốt (70)')
ax.axvline(50, color='orange', linestyle='--', linewidth=2.5, alpha=0.7, label='Ngưỡng TB (50)')

# Thêm số điểm trên mỗi bar
for i, (bar, score) in enumerate(zip(bars, top40['health'])):
    ax.text(score + 1, i, f'{score:.0f}', va='center', fontsize=10, fontweight='bold')

ax.legend(loc='lower right', fontsize=13)
ax.grid(axis='x', alpha=0.3, linestyle='--')
ax.set_xlim(0, 105)

plt.tight_layout()
plt.show()

# BIỂU ĐỒ 3: P/E vs DIVIDEND YIELD (Định giá)

fig, ax = plt.subplots(figsize=(16, 12))

# Lọc dữ liệu
valid = df[(df['price_earnings'] > 0) & (df['price_earnings'] < 80)]

# Scatter plot
scatter = ax.scatter(
    valid['price_earnings'], 
    valid['dividend_yield'],
    c=valid['market_cap_log__scaled'],
    s=120,
    alpha=0.6,
    cmap='viridis',
    edgecolors='black',
    linewidths=0.5
)

# Vẽ vùng phân loại
ax.axvspan(0, 15, alpha=0.1, color='blue', label='Value Stocks (P/E < 15)')
ax.axvspan(15, 25, alpha=0.1, color='green', label='Fair Value (P/E 15-25)')
ax.axvspan(25, 80, alpha=0.1, color='orange', label='Growth Stocks (P/E > 25)')
ax.axhline(2, color='red', linestyle='--', linewidth=2, alpha=0.5, label='Dividend 2% (ngưỡng tốt)')

# Labels
ax.set_xlabel('P/E Ratio (Price-to-Earnings)', fontsize=16, fontweight='bold')
ax.set_ylabel('Dividend Yield (%)', fontsize=16, fontweight='bold')
ax.set_title('QUAN HỆ ĐỊNH GIÁ: P/E vs DIVIDEND YIELD\n(Màu sắc = Quy mô công ty)', 
             fontsize=18, fontweight='bold', pad=20)

# Annotations
ax.text(0.02, 0.98, '💰 VALUE STOCKS\nP/E thấp, Dividend cao\n(Ổn định, thu nhập)', 
        transform=ax.transAxes, fontsize=13, verticalalignment='top',
        bbox=dict(boxstyle='round,pad=0.8', facecolor='lightblue', alpha=0.8, edgecolor='blue', linewidth=2),
        fontweight='bold')

ax.text(0.98, 0.02, '🚀 GROWTH STOCKS\nP/E cao, Dividend thấp\n(Tăng trưởng, rủi ro)', 
        transform=ax.transAxes, fontsize=13, verticalalignment='bottom', horizontalalignment='right',
        bbox=dict(boxstyle='round,pad=0.8', facecolor='lightyellow', alpha=0.8, edgecolor='orange', linewidth=2),
        fontweight='bold')

# Colorbar
cbar = plt.colorbar(scatter, ax=ax, pad=0.02)
cbar.set_label('Market Cap (log scaled)', fontsize=14, fontweight='bold')

ax.legend(loc='upper right', fontsize=12)
ax.grid(True, alpha=0.3, linestyle='--')

plt.tight_layout()
plt.show()

# BIỂU ĐỒ 4: PHÂN PHỐI CÁC CHỈ SỐ QUAN TRỌNG (Dashboard 4 biểu đồ)

fig = plt.figure(figsize=(18, 12))

# 4.1 P/E Ratio Distribution
ax1 = plt.subplot(2, 2, 1)
pe = df[(df['price_earnings'] > 0) & (df['price_earnings'] < 100)]['price_earnings']
ax1.hist(pe, bins=40, color='steelblue', edgecolor='black', alpha=0.7, linewidth=1.2)
ax1.axvline(pe.median(), color='red', linestyle='--', linewidth=3, label=f'Median: {pe.median():.1f}')
ax1.axvline(pe.mean(), color='green', linestyle='--', linewidth=3, label=f'Mean: {pe.mean():.1f}')
ax1.axvline(15, color='blue', linestyle=':', linewidth=2, alpha=0.7, label='Value (15)')
ax1.axvline(25, color='orange', linestyle=':', linewidth=2, alpha=0.7, label='Growth (25)')
ax1.set_xlabel('P/E Ratio', fontsize=14, fontweight='bold')
ax1.set_ylabel('Số lượng công ty', fontsize=14, fontweight='bold')
ax1.set_title('Phân phối P/E Ratio', fontsize=16, fontweight='bold', pad=15)
ax1.legend(fontsize=11)
ax1.grid(alpha=0.3, linestyle='--')

# 4.2 Market Cap Distribution
ax2 = plt.subplot(2, 2, 2)
def cat_cap(x):
    b = x / 1e9
    if b > 200: return 'Mega Cap\n(>$200B)'
    elif b >= 10: return 'Large Cap\n($10-200B)'
    elif b >= 2: return 'Mid Cap\n($2-10B)'
    else: return 'Small Cap\n(<$2B)'

df['cap_cat'] = df['market_cap'].apply(cat_cap)
cap_c = df['cap_cat'].value_counts()
colors = ['#8b5cf6', '#3b82f6', '#10b981', '#f59e0b']
wedges, texts, autotexts = ax2.pie(cap_c, labels=cap_c.index, autopct='%1.1f%%', 
                                     startangle=90, colors=colors, textprops={'fontsize': 12, 'weight': 'bold'})
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(13)
    autotext.set_weight('bold')
ax2.set_title('Phân loại theo Vốn hóa', fontsize=16, fontweight='bold', pad=15)

# 4.3 Dividend Distribution
ax3 = plt.subplot(2, 2, 3)
has_div = df[df['dividend_yield'] > 0]
no_div = df[df['dividend_yield'] == 0]
data = [len(has_div), len(no_div)]
labels = [f'Có trả cổ tức\n({len(has_div)} công ty)', f'Không trả cổ tức\n({len(no_div)} công ty)']
colors_div = ['#10b981', '#ef4444']
wedges, texts, autotexts = ax3.pie(data, labels=labels, autopct='%1.1f%%', 
                                     startangle=90, colors=colors_div, textprops={'fontsize': 12, 'weight': 'bold'})
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(14)
    autotext.set_weight('bold')
ax3.set_title('Chính sách Cổ tức', fontsize=16, fontweight='bold', pad=15)

# 4.4 Health Score Distribution
ax4 = plt.subplot(2, 2, 4)
ax4.hist(df['health'], bins=25, color='royalblue', edgecolor='black', alpha=0.7, linewidth=1.2)
ax4.axvline(df['health'].median(), color='red', linestyle='--', linewidth=3,
            label=f'Median: {df["health"].median():.1f}')
ax4.axvline(70, color='green', linestyle='--', linewidth=2.5, alpha=0.7, label='Tốt (70)')
ax4.axvline(50, color='orange', linestyle='--', linewidth=2.5, alpha=0.7, label='TB (50)')
ax4.set_xlabel('Health Score (0-100)', fontsize=14, fontweight='bold')
ax4.set_ylabel('Số lượng công ty', fontsize=14, fontweight='bold')
ax4.set_title('Điểm Sức khỏe Tài chính', fontsize=16, fontweight='bold', pad=15)
ax4.legend(fontsize=11)
ax4.grid(alpha=0.3, linestyle='--')

plt.suptitle('DASHBOARD PHÂN PHỐI CÁC CHỈ SỐ QUAN TRỌNG', 
             fontsize=20, fontweight='bold', y=0.995)
plt.tight_layout()
plt.show()

# BIỂU ĐỒ 5: MA TRẬN TƯƠNG QUAN (Heatmap)

fig, ax = plt.subplots(figsize=(14, 11))

# Chọn các chỉ số quan trọng
corr_cols = ['price_earnings', 'price_book', 'price_sales', 'dividend_yield', 
             'profitability_ratio__scaled', 'market_cap_log__scaled', 
             'price_to_52w_high__scaled', 'ebitda_log__scaled']

corr_labels = ['P/E Ratio', 'P/B Ratio', 'P/S Ratio', 'Dividend Yield',
               'Profitability', 'Market Cap (log)', 'Volatility', 'EBITDA (log)']

corr_matrix = df[corr_cols].corr()

# Vẽ heatmap
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0, 
            ax=ax, cbar_kws={'shrink': 0.8}, linewidths=2, linecolor='white',
            square=True, annot_kws={'size': 13, 'weight': 'bold'})

# Đặt labels
ax.set_xticklabels(corr_labels, rotation=45, ha='right', fontsize=12)
ax.set_yticklabels(corr_labels, rotation=0, fontsize=12)
ax.set_title('MA TRẬN TƯƠNG QUAN CÁC CHỈ SỐ TÀI CHÍNH\n(Giá trị càng gần ±1 = tương quan càng mạnh)', 
             fontsize=18, fontweight='bold', pad=20)

plt.tight_layout()
plt.show()

# BIỂU ĐỒ 6: PHÂN LOẠI RISK-RETURN (Quadrant Analysis)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8))

# Tính quadrant
rr = df[['price_to_52w_high__scaled', 'profitability_ratio__scaled', 'market_cap']].dropna()

def classify(row):
    r = row['price_to_52w_high__scaled']
    ret = row['profitability_ratio__scaled']
    if ret > 0 and r < 0: return '🎯 Tốt\n(High Return,\nLow Risk)'
    elif ret > 0 and r >= 0: return '📈 Trung bình+\n(High Return,\nHigh Risk)'
    elif ret <= 0 and r < 0: return '💼 Trung bình-\n(Low Return,\nLow Risk)'
    else: return '⚠️ Kém\n(Low Return,\nHigh Risk)'

rr['quadrant'] = rr.apply(classify, axis=1)
quad_counts = rr['quadrant'].value_counts()

# Pie chart
colors_quad = ['#10b981', '#3b82f6', '#f59e0b', '#ef4444']
wedges, texts, autotexts = ax1.pie(quad_counts, labels=quad_counts.index, autopct='%1.1f%%',
                                     startangle=90, colors=colors_quad, 
                                     textprops={'fontsize': 13, 'weight': 'bold'},
                                     explode=[0.05, 0, 0, 0.05])
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(14)
    autotext.set_weight('bold')

ax1.set_title('Phân loại Risk-Return', fontsize=16, fontweight='bold', pad=15)

# Bar chart
bars = ax2.bar(range(len(quad_counts)), quad_counts.values, color=colors_quad, 
               edgecolor='black', linewidth=1.5, alpha=0.8)
ax2.set_xticks(range(len(quad_counts)))
ax2.set_xticklabels([q.replace('\n', ' ') for q in quad_counts.index], 
                     rotation=15, ha='right', fontsize=11)
ax2.set_ylabel('Số lượng công ty', fontsize=14, fontweight='bold')
ax2.set_title('Số lượng công ty theo từng nhóm', fontsize=16, fontweight='bold', pad=15)
ax2.grid(axis='y', alpha=0.3, linestyle='--')

# Thêm số lượng trên mỗi bar
for bar, count in zip(bars, quad_counts.values):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + 1,
            f'{count}\n({count/len(rr)*100:.1f}%)',
            ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.suptitle('PHÂN LOẠI CÔNG TY THEO RỦI RO - LỢI NHUẬN', 
             fontsize=20, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```
