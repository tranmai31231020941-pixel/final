```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Đọc dữ liệu
df = pd.read_csv('financials_features_scaled.csv')

# Thiết lập style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")

# ============================================================
# BIỂU ĐỒ 1: PHÂN BỐ CHỈ SỐ ĐỊNH GIÁ
# ============================================================
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
valuation_cols = ['price_earnings', 'price_book', 'price_sales']
titles = ['P/E Ratio', 'P/B Ratio', 'P/S Ratio']
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']

for idx, (col, title, color) in enumerate(zip(valuation_cols, titles, colors)):
    data_filtered = df[df[col] < df[col].quantile(0.95)]
    
    parts = axes[idx].violinplot([data_filtered[data_filtered['dividend_flag__scaled'] == 0][col].dropna(),
                                   data_filtered[data_filtered['dividend_flag__scaled'] == -1][col].dropna()],
                                  positions=[1, 2],
                                  showmeans=True,
                                  showmedians=True)
    
    for pc in parts['bodies']:
        pc.set_facecolor(color)
        pc.set_alpha(0.7)
    
    axes[idx].set_title(title, fontsize=14, fontweight='bold')
    axes[idx].set_xticks([1, 2])
    axes[idx].set_xticklabels(['Có cổ tức', 'Không cổ tức'])
    axes[idx].grid(True, alpha=0.3, axis='y')

plt.suptitle('PHÂN BỐ ĐỊNH GIÁ: SO SÁNH CÔNG TY CÓ/KHÔNG CỔ TỨC', 
             fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# ============================================================
# BIỂU ĐỒ 2: VỐN HÓA VS KHẢ NĂNG SINH LỜI
# ============================================================
fig = plt.figure(figsize=(16, 10))
ax = fig.add_subplot(111)

# Chuẩn bị dữ liệu
plot_data = df[['market_cap_log__scaled', 'profitability_ratio__scaled', 
                'valuation_z__scaled', 'dividend_yield', 'dividend_flag__scaled']].dropna()

# Tạo bubble chart
scatter = ax.scatter(plot_data['market_cap_log__scaled'],
                     plot_data['profitability_ratio__scaled'],
                     s=np.abs(plot_data['valuation_z__scaled']) * 100 + 20,
                     c=plot_data['dividend_yield'],
                     cmap='viridis',
                     alpha=0.6,
                     edgecolors='black',
                     linewidth=0.5)

# Thêm đường phân vùng
ax.axhline(y=0, color='red', linestyle='--', alpha=0.5, linewidth=2, label='Sinh lời = 0')
ax.axvline(x=0, color='blue', linestyle='--', alpha=0.5, linewidth=2, label='Vốn hóa TB')

# Tô màu các vùng
ax.fill_between([-2, 3], 0, 5, alpha=0.1, color='green', label='Vùng tốt: Lớn & sinh lời cao')
ax.fill_between([-2, 3], -5, 0, alpha=0.1, color='red', label='Vùng rủi ro')

ax.set_xlabel('Quy mô vốn hóa (log scaled)', fontsize=14, fontweight='bold')
ax.set_ylabel('Khả năng sinh lời', fontsize=14, fontweight='bold')
ax.set_title('PHÂN LOẠI CÔNG TY: VỐN HÓA × SINH LỜI × ĐỊNH GIÁ\n(Kích thước bong bóng = Mức định giá)', 
             fontsize=16, fontweight='bold', pad=20)

cbar = plt.colorbar(scatter, ax=ax)
cbar.set_label('Tỷ suất cổ tức (%)', fontsize=12, fontweight='bold')
ax.legend(loc='upper left', fontsize=10)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# ============================================================
# BIỂU ĐỒ 3: PHÂN PHỐI EBITDA
# ============================================================
plt.figure(figsize=(14, 8))
ebitda_positive = df[df['ebitda_log__scaled'] > -5]['ebitda_log__scaled']

plt.hist(ebitda_positive, bins=60, color='#2ECC71', edgecolor='black', 
         alpha=0.7, linewidth=1.2)
plt.axvline(ebitda_positive.mean(), color='red', linestyle='--', 
            linewidth=2, label=f'Trung bình: {ebitda_positive.mean():.2f}')
plt.axvline(ebitda_positive.median(), color='blue', linestyle='--', 
            linewidth=2, label=f'Trung vị: {ebitda_positive.median():.2f}')

plt.xlabel('EBITDA (log scaled)', fontsize=14, fontweight='bold')
plt.ylabel('Số lượng công ty', fontsize=14, fontweight='bold')
plt.title('PHÂN PHỐI EBITDA - HIỆU QUẢ HOẠT ĐỘNG', fontsize=18, fontweight='bold', pad=20)
plt.legend(fontsize=12, loc='best')
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

# BIỂU ĐỒ 4: SO SÁNH CÔNG TY CÓ/KHÔNG CỔ TỨC
# ============================================================
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

# Chart 1: So sánh các chỉ số trung bình
dividend_comparison = df.groupby('dividend_flag__scaled')[
    ['profitability_ratio__scaled', 'market_cap_log__scaled', 'valuation_z__scaled']
].mean()

x = np.arange(len(dividend_comparison.columns))
width = 0.35
labels = ['Không cổ tức (-1)', 'Có cổ tức (0)']
colors_bars = ['#E74C3C', '#3498DB']

for idx, (label, color) in enumerate(zip(labels, colors_bars)):
    values = dividend_comparison.iloc[idx].values
    ax1.bar(x + idx*width, values, width, label=label, color=color, alpha=0.8)

ax1.set_xlabel('Chỉ số tài chính', fontsize=12, fontweight='bold')
ax1.set_ylabel('Giá trị trung bình', fontsize=12, fontweight='bold')
ax1.set_title('So sánh chỉ số trung bình', fontsize=14, fontweight='bold')
ax1.set_xticks(x + width/2)
ax1.set_xticklabels(['Sinh lời', 'Vốn hóa', 'Định giá'])
ax1.legend()
ax1.grid(True, alpha=0.3, axis='y')

# Chart 2: Số lượng công ty
dividend_counts = df['dividend_flag__scaled'].value_counts()
colors_pie = ['#3498DB', '#E74C3C']
labels_pie = [f'Có cổ tức\n({dividend_counts[0]} công ty)', 
              f'Không cổ tức\n({dividend_counts[-1]} công ty)']

ax2.pie(dividend_counts.values, labels=labels_pie, colors=colors_pie,
        autopct='%1.1f%%', startangle=90, textprops={'fontsize': 12, 'fontweight': 'bold'})
ax2.set_title('Tỷ lệ công ty có/không cổ tức', fontsize=14, fontweight='bold')

plt.suptitle('PHÂN TÍCH CÔNG TY CÓ/KHÔNG CỔ TỨC', 
             fontsize=18, fontweight='bold', y=0.98)
plt.tight_layout()
plt.savefig('5_dividend_comparison.png', dpi=300, bbox_inches='tight')
plt.show()
```
